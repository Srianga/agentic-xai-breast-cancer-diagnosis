```python
import matplotlib.pyplot as plt

metrics = ['Accuracy', 'Precision', 'Recall', 'F1']
values = [0.95, 0.94, 0.93, 0.94]

plt.plot(metrics, values, marker='o')
plt.title("Model Performance Metrics")
plt.xlabel("Metrics")
plt.ylabel("Score")

plt.savefig("performance.png")
plt.show()
```
